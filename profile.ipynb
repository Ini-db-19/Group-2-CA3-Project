```python
import tkinter as tk
from PIL import Image, ImageTk
from tkinter import messagebox

def exit_app(event=None):
    root.destroy()

root = tk.Tk()
root.attributes("-fullscreen", True)
root.configure(bg='#EDE3D6')
root.bind('<Escape>', exit_app)

# Header frame
header_bg_frame = tk.Frame(root, bg='#FFFFFF', height=80)
header_bg_frame.place(x=0, y=0, width=root.winfo_screenwidth(), height=80)
profile_title = tk.Label(header_bg_frame, text="Personal Profile", font=('Segoe UI', 28, 'bold'),
                         fg='#1F4E79', bg='#FFFFFF')
profile_title.place(relx=0.55, rely=0.5, anchor='center')

# Main content
x_offset = 280
content_width = root.winfo_screenwidth() - x_offset - 40
screen_height = root.winfo_screenheight()

# info frame
info_width = 350                      
info_height = 140                      
info_y = 130                          

info_x = x_offset + (content_width - info_width) // 2 - 40
info_frame = tk.Frame(root, bd=2, relief="solid", bg='#EDE3D6')
info_frame.place(x=info_x, y=info_y, width=info_width, height=info_height)

#Image of profile pic
image_path = "assets/prof_pic.png"
pil_img = Image.open(image_path)
pil_img = pil_img.resize((80, 80))    # image size
photo_img = ImageTk.PhotoImage(pil_img)


img_label = tk.Label(info_frame, image=photo_img, bg='#EDE3D6')
img_label.image = photo_img
img_label.place(x=5, y=20)         

# Text in the profile frame
text_frame = tk.Frame(info_frame, bg='#EDE3D6')
text_frame.place(x=90, y=8)          


tk.Label(text_frame, text="Username: John Doe", font=('Segoe UI', 11),
         bg='#EDE3D6', fg='#2C3E50').pack(anchor='w', pady=1)
tk.Label(text_frame, text="Location: Lagos, Nigeria", font=('Segoe UI', 11),
         bg='#EDE3D6', fg='#2C3E50').pack(anchor='w', pady=1)
tk.Label(text_frame, text="Ratings: 4.8 ★", font=('Segoe UI', 11),
         bg='#EDE3D6', fg='#2C3E50').pack(anchor='w', pady=1)
tk.Label(text_frame, text="Money earned: ₦125,000", font=('Segoe UI', 11),
         bg='#EDE3D6', fg='#2C3E50').pack(anchor='w', pady=1)


combined_y = info_y + info_height + 20 

# Combined frame (Experience, Skills & Certifications)
combined_height = 210
combined_y = screen_height // 2 - combined_height // 2 + 40

combined_frame = tk.Frame(root, bd=2, relief="solid", bg='#EDE3D6')
combined_frame.place(x=x_offset, y=combined_y, width=content_width, height=combined_height)

# Experience
exp_frame = tk.Frame(combined_frame, bg='#EDE3D6')
exp_frame.place(x=2, y=2, width=content_width-8, height=105)
exp_label = tk.Label(exp_frame, text="Experience", font=('Segoe UI', 14, 'bold'),
                     bg='#EDE3D6', fg='#2C3E50')
exp_label.place(x=20, y=15)
exp_desc = tk.Label(exp_frame, text="Add your experience to get tailored job recommendations.",
                    font=('Segoe UI', 11), bg='#EDE3D6', fg='#555')
exp_desc.place(x=20, y=55)

# Skills & Languages and Certifications
half_width = content_width // 2

skills_frame = tk.Frame(combined_frame, bg='#EDE3D6', width=half_width, height=85)
skills_frame.place(x=2, y=102, width=(content_width-8)//2, height=85)   # y was 109, now 102
skills_label = tk.Label(skills_frame, text="Skills & Languages", font=('Segoe UI', 14, 'bold'),
                        bg='#EDE3D6', fg='#2C3E50')
skills_label.place(x=20, y=8)
skills_btn = tk.Button(skills_frame, text="Add a skill or language", font=('Segoe UI', 10),
                       bg='#3498DB', fg='white', bd=0, padx=10, pady=5, cursor="hand2",
                       command=lambda: messagebox.showinfo("Add Skill", "Feature coming soon"))
skills_btn.place(x=20, y=45)   # adjusted y from 50 to 45

cert_frame = tk.Frame(combined_frame, bg='#EDE3D6', width=half_width, height=85)
cert_frame.place(x=2 + (content_width-8)//2, y=102, width=(content_width-8)//2, height=85)
cert_label = tk.Label(cert_frame, text="Certifications & License", font=('Segoe UI', 14, 'bold'),
                      bg='#EDE3D6', fg='#2C3E50')
cert_label.place(x=20, y=8)
cert_btn = tk.Button(cert_frame, text="Add a certificate or license", font=('Segoe UI', 10),
                     bg='#3498DB', fg='white', bd=0, padx=10, pady=5, cursor="hand2",
                     command=lambda: messagebox.showinfo("Add Certificate", "Feature coming soon"))
cert_btn.place(x=20, y=45)

# Job preferences
pref_y = combined_y + combined_height + 40
pref_height = 180

pref_frame = tk.Frame(root, bd=2, relief="solid", bg='#EDE3D6')
pref_frame.place(x=x_offset, y=pref_y, width=content_width, height=pref_height)

pref_label = tk.Label(pref_frame, text="Job Preferences", font=('Segoe UI', 14, 'bold'),
                      bg='#EDE3D6', fg='#2C3E50')
pref_label.place(x=20, y=15)
pref_desc = tk.Label(pref_frame, text="Get customized job recommendations by keeping your preferences updated",
                     font=('Segoe UI', 11), bg='#EDE3D6', fg='#555')
pref_desc.place(x=20, y=55)

# Buttons 
btn_frame = tk.Frame(pref_frame, bg='#EDE3D6')
btn_frame.place(x=20, y=100)

btn1 = tk.Button(btn_frame, text="Desired Job Title", font=('Segoe UI', 10),
                 bg='#ECF0F1', fg='#2C3E50', bd=1, padx=10, pady=5, cursor="hand2",
                 command=lambda: messagebox.showinfo("Job Title", "Set desired job title"))
btn1.pack(side='left', padx=5)

btn2 = tk.Button(btn_frame, text="Desired Work Location", font=('Segoe UI', 10),
                 bg='#ECF0F1', fg='#2C3E50', bd=1, padx=10, pady=5, cursor="hand2",
                 command=lambda: messagebox.showinfo("Location", "Set work location"))
btn2.pack(side='left', padx=5)

btn3 = tk.Button(btn_frame, text="Desired Minimum Pay", font=('Segoe UI', 10),
                 bg='#ECF0F1', fg='#2C3E50', bd=1, padx=10, pady=5, cursor="hand2",
                 command=lambda: messagebox.showinfo("Minimum Pay", "Set desired pay"))
btn3.pack(side='left', padx=5)

btn4 = tk.Button(btn_frame, text="Remote work preferences", font=('Segoe UI', 10),
                 bg='#ECF0F1', fg='#2C3E50', bd=1, padx=10, pady=5, cursor="hand2",
                 command=lambda: messagebox.showinfo("Remote Work", "Set remote preference"))
btn4.pack(side='left', padx=5)


# Sidebar
sidebar = tk.Frame(root, bg='#1E1E2E', width=250, relief='flat')
sidebar.place(x=0, y=0, width=250, height=root.winfo_screenheight())
sidebar.pack_propagate(False)

header_frame = tk.Frame(sidebar, bg='#1E1E2E', height=100)
header_frame.pack(fill='x', pady=(20, 10))
logo_text = tk.Label(header_frame, text="Hustle Hub", font=('Segoe UI', 18, 'bold'),
                     fg='#FFB347', bg='#1E1E2E')
logo_text.pack(pady=10)
separator = tk.Frame(sidebar, bg='#3A3A4A', height=2)
separator.pack(fill='x', padx=20, pady=5)

def make_sidebar_button(text, icon, msg_title, msg_text):
    frame = tk.Frame(sidebar, bg='#1E1E2E')
    frame.pack(fill='x', padx=15, pady=5)
    label = tk.Label(frame, text=f"{icon}  {text}", font=('Segoe UI', 12),
                     fg='white', bg='#1E1E2E', anchor='w', cursor="hand2")
    label.pack(fill='x', padx=10, pady=8)
    label.bind('<Button-1>', lambda e: messagebox.showinfo(msg_title, msg_text))

make_sidebar_button("Home", "🏠", "Home", "Welcome to your home feed!")
make_sidebar_button("My Jobs", "💼", "My Jobs", "View and manage your job applications.")
make_sidebar_button("Messages", "💬", "Messages", "Check your conversations.")
make_sidebar_button("Profile", "👤", "Profile", "Your personal information and settings.")

bottom_frame = tk.Frame(sidebar, bg='#1E1E2E')
bottom_frame.pack(side='bottom', fill='x', pady=20)
logout_btn = tk.Button(bottom_frame, text="Logout", font=('Segoe UI', 10),
                       bg='#FF6B6B', fg='white', bd=0, padx=10, pady=5,
                       cursor="hand2", command=lambda: messagebox.showinfo("Logout", "You have been logged out."))
logout_btn.pack(pady=5)

root.mainloop()
```
