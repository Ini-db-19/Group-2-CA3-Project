```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

def slide_image_down():
    global y_pos
    if y_pos < root.winfo_height():
        y_pos += step
        sliding_label.place(x=0, y=y_pos, relwidth=1, relheight=1)
        root.after(delay, slide_image_down)
    else:
        logo_label.place(relx=0, x=100, rely=0, y=40)
        hello_label.place(relx=0.5, rely=0.25, anchor='center')
        welcome_label.place(relx=0.5, rely=0.35, anchor='center')
        tagline_label.place(relx=0.5, rely=0.72, anchor='center')
        next_button.place(relx=0.5, rely=0.95, y=-50, anchor='s')

def exit_app(event=None):
    root.destroy()

def on_button_click():
    messagebox.showinfo("Info", "Login page will open soon.")

# Create window
root = tk.Tk()
root.title("Hustle Hub")
root.attributes('-fullscreen', True)
root.bind('<Escape>', exit_app)
root.update()

# Background color (what appears as image slides away)
root.configure(bg='#B05938')   # or any color/image

# Load full-screen image
img = Image.open("assets/bg_image.png")
img = img.resize((root.winfo_width(), root.winfo_height()))
photo = ImageTk.PhotoImage(img)

# Label that starts at top-left (fully visible)
sliding_label = tk.Label(root, image=photo)
sliding_label.image = photo
sliding_label.place(x=0, y=0, relwidth=1, relheight=1)

# Animation parameters
y_pos = 0
step = 10      # pixels per frame – larger = faster
delay = 10     # milliseconds per frame

# Start sliding down after 1 second (adjust as needed)
root.after(1000, slide_image_down)

logo_img = Image.open("assets/logo.png")
logo_img = logo_img.resize((300, 299))
logo_photo = ImageTk.PhotoImage(logo_img)
logo_label = tk.Label(root, image=logo_photo, bg='#B05938')
logo_label.image = logo_photo

hello_label = tk.Label(root, text="Hey there!", font=('Segoe UI', 40, 'bold'), fg='#F8F0E3', bg='#B05938')

welcome_label = tk.Label(root, text="Welcome to Hustle Hub", font=('Segoe UI', 28, 'bold'), fg='#F8F0E3', bg='#B05938')

# Tagline
tagline_label = tk.Label(root, text="Turn your hustle into money now", font=('Segoe Script', 20, 'italic'), fg='#FFE0C0', bg='#B05938')


next_button = tk.Button(root, text="Get Started", font=('Arial', 25, 'bold'),
                        bg='#E4D1B9', fg='#B05938', padx=30, pady=10,
                        cursor='hand2', bd=0, command=on_button_click)

root.mainloop()
```
