```python
import tkinter as tk
from tkinter import messagebox

def exit_app():
    root.destroy


def search_jobs():
    term = search_entry.get()
    if term:
        messagebox.showinfo("Search", f"Searching for: {term}")
    else:
        messagebox.showwarning("Search", "Enter a job title or location")


def apply_for_job(title):
    confirm = messagebox.askyesno("Apply", f"Apply for {title}?")
    if confirm:
        messagebox.showinfo("Applied", f"You applied for {title}!")


def exit_app(event=None):
    root.destroy()


# Main window
root = tk.Tk()
root.title("HustleHub")
root.attributes('-fullscreen', True)
root.bind('<Escape>', exit_app)
root.configure(bg='#EDE3D6')


# Get screen width and height
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()


# SIDEBAR 
sidebar = tk.Frame(root, bg='#FFFFFF', width=250, height=screen_height, relief='flat')
sidebar.place(x=0, y=0)
sidebar.pack_propagate(False)


header_frame = tk.Frame(sidebar, bg='#FFFFFF', height=100)
header_frame.pack(fill='x', pady=(20, 10))
logo_text = tk.Label(header_frame, text="Hustle Hub", font=('Segoe UI', 18, 'bold'),
                     fg='#FFB347', bg='#FFFFFF')
logo_text.pack(pady=10)
separator = tk.Frame(sidebar, bg='#E0E0E0', height=2)
separator.pack(fill='x', padx=20, pady=5)


def make_sidebar_button(text, icon, msg_title, msg_text):
    frame = tk.Frame(sidebar, bg='#FFFFFF')
    frame.pack(fill='x', padx=15, pady=5)
    label = tk.Label(frame, text=f"{icon}  {text}", font=('Segoe UI', 12),
                     fg='#333333', bg='#FFFFFF', anchor='w', cursor="hand2")
    label.pack(fill='x', padx=10, pady=8)
    label.bind('<Button-1>', lambda e: messagebox.showinfo(msg_title, msg_text))


make_sidebar_button("Home", "🏠", "Home", "Welcome to your home feed!")
make_sidebar_button("My Jobs", "💼", "My Jobs", "View and manage your job applications.")
make_sidebar_button("Messages", "💬", "Messages", "Check your conversations.")
make_sidebar_button("Profile", "👤", "Profile", "Your personal information and settings.")


bottom_frame = tk.Frame(sidebar, bg='#FFFFFF')
bottom_frame.pack(side='bottom', fill='x', pady=20)
logout_btn = tk.Button(bottom_frame, text="Logout", font=('Segoe UI', 10),
                       bg='#FF6B6B', fg='white', bd=0, padx=10, pady=5,
                       cursor="hand2", command=lambda: messagebox.showinfo("Logout", "You have been logged out."))
logout_btn.pack(pady=5)


# MAIN CONTENT AREA
main_area = tk.Frame(root, bg='#EDE3D6')
main_area.place(x=250, y=0, width=screen_width-250, height=screen_height)


# ========== SEARCH SECTION ==========
search_container = tk.Frame(main_area, bg='#EDE3D6')
search_container.pack(fill='x', pady=40)


search_center = tk.Frame(search_container, bg='#EDE3D6')
search_center.pack()


# Create a frame to hold search icon and entry side by side
search_row = tk.Frame(search_center, bg='#EDE3D6')
search_row.pack()


# Search icon (⌕) as a label
search_icon = tk.Label(search_row, text="⌕", font=('Arial', 20),
                       fg='#B85C2E', bg='#EDE3D6')
search_icon.pack(side='left', padx=(0, 10))


# Search entry
search_entry = tk.Entry(search_row, font=('Arial', 16), width=50, bd=2, relief='solid')
search_entry.pack(side='left')


# Bind Enter key to search
search_entry.bind('<Return>', lambda event: search_jobs())


# Decorative text below search bar (not clickable, just for aesthetics)
decorative_text = tk.Label(search_center, text="Browse available Jobs", 
                           font=('Segoe UI', 12, 'italic'),
                           fg='#B85C2E', bg='#EDE3D6')
decorative_text.pack(pady=(15, 0))


# SCROLLABLE JOB LISTINGS 
listings_container = tk.Frame(main_area, bg='#EDE3D6')
listings_container.pack(fill='both', expand=True, padx=40, pady=(20, 20))


canvas = tk.Canvas(listings_container, bg='#EDE3D6', highlightthickness=0)
scrollbar = tk.Scrollbar(listings_container, orient='vertical', command=canvas.yview)
canvas.configure(yscrollcommand=scrollbar.set)


scrollbar.pack(side='right', fill='y')
canvas.pack(side='left', fill='both', expand=True)


jobs_frame = tk.Frame(canvas, bg='#EDE3D6')
canvas_window = canvas.create_window((0, 0), window=jobs_frame, anchor='nw')


def center_jobs(event):
    canvas_width = event.width
    frame_width = jobs_frame.winfo_reqwidth()
    if frame_width > 0 and canvas_width > frame_width:
        x = (canvas_width - frame_width) // 2
        canvas.coords(canvas_window, x, 0)
    else:
        canvas.coords(canvas_window, 0, 0)


def update_scroll_region(event):
    canvas.configure(scrollregion=canvas.bbox('all'))


def on_mousewheel(event):
    canvas.yview_scroll(int(-1*(event.delta/120)), 'units')


jobs_frame.bind('<Configure>', update_scroll_region)
canvas.bind('<Configure>', center_jobs)
canvas.bind('<MouseWheel>', on_mousewheel)


# JOB CARD FUNCTION 
def add_job(title, company, experience, address, time, salary):
    card = tk.Frame(jobs_frame, bg='white', relief='solid', bd=1)
    card.pack(fill='x', pady=10, padx=10)
    
    # Title row
    top_row = tk.Frame(card, bg='white')
    top_row.pack(fill='x', padx=15, pady=(12, 5))
    
    title_label = tk.Label(top_row, text=title, font=('Arial', 18, 'bold'), 
                            fg='#FFB347', bg='white')
    title_label.pack(side='left')
    
    plus_btn = tk.Button(top_row, text="+", font=('Arial', 20, 'bold'), 
                          bg='#FFB347', fg='#1E1E2E', bd=0, width=3,
                          cursor='hand2', command=lambda: apply_for_job(title))
    plus_btn.pack(side='right')
    
    # Company
    tk.Label(card, text=company, font=('Arial', 14), fg='#333333', bg='white')\
        .pack(anchor='w', padx=15, pady=2)
    
    # Experience
    tk.Label(card, text=f"• {experience}", font=('Arial', 12), fg='#666666', bg='white')\
        .pack(anchor='w', padx=15, pady=2)
    
    # Address
    tk.Label(card, text=f"• {address}", font=('Arial', 12), fg='#666666', bg='white')\
        .pack(anchor='w', padx=15, pady=2)
    
    # Bottom row
    bottom_row = tk.Frame(card, bg='white')
    bottom_row.pack(fill='x', padx=15, pady=(5, 12))
    
    tk.Label(bottom_row, text=time, font=('Arial', 11), fg='#888888', bg='white')\
        .pack(side='left')
    
    if salary:
        tk.Label(bottom_row, text=salary, font=('Arial', 12, 'bold'), 
                 fg='#FFB347', bg='white').pack(side='right')


# JOB LISTINGS 
add_job("Plumber", "Sola's Shack", "At least 2 years experience needed",
        "2. Akimpemri Cres., Mushin, Lagos", "6 hours ago", "₦70,000")


add_job("Cleaner", "Royal Ebony College", "None needed",
        "Radere, Ajah, Lekki-Epe Expressway, Lagos", "8 hours ago", "₦55,000")


add_job("Carpenter", "Dr Clinton Obasanjo", "At least 5 years experience needed",
        "No. 27 Emerald Avenue, VUC, Lekki, Lagos", "25 minutes ago", "₦85,000")


add_job("Mechanic", "AutoCare Center", "At least 3 years experience needed",
        "15, Ikorodu Road, Maryland, Lagos", "2 hours ago", "₦80,000")


add_job("Chef", "Tasty Bites Restaurant", "At least 2 years experience needed",
        "42, Admiralty Way, Lekki Phase 1, Lagos", "1 day ago", "₦95,000")


add_job("Electrician", "PowerFix Solutions", "At least 2 years experience needed",
        "23, Allen Avenue, Ikeja, Lagos", "3 hours ago", "₦65,000")


add_job("Driver", "LogiStar Transport", "Valid drivers license required",
        "Plot 5, Mobolaji Bank-Anthony Way, Ikeja", "12 hours ago", "₦75,000")


root.mainloop()
```
