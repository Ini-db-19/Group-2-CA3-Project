```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk



# ==============================================================================
# FUNCTIONS
# ==============================================================================
def on_employer_click():
    messagebox.showinfo(
        "Employer",
        "You selected: I want to hire workers for my company."
    )

def on_employee_click():
    messagebox.showinfo(
        "Employee",
        "You selected: I am looking for jobs and opportunities."
    )

def go_to_signin():
    messagebox.showinfo("Sign In", "Proceeding to the sign-in page...")

def exit_app(event=None):
    root.destroy()

# ==============================================================================
# MAIN WINDOW SETUP
# ==============================================================================
root = tk.Tk()
root.title("HustleHub")

# Windowed fullscreen format optimized for Jupyter stability on Mac
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
root.attributes('-fullscreen',True)

root.bind('<Escape>', exit_app)
root.configure(bg='white')

# ==============================================================================
# 1. BACKGROUND LAYER (Must be drawn first)
# ==============================================================================
bg_image = Image.open("assets/bg_image.png")
bg_image = bg_image.resize((screen_width, screen_height))
bg_photo = ImageTk.PhotoImage(bg_image)

root.bg_photo = bg_photo  # Prevent garbage collection

bg_label = tk.Label(root, image=root.bg_photo)
bg_label.place(x=0, y=0, relwidth=1, relheight=1)

# ==============================================================================
# 2. LOGO LAYER (Child of bg_label so it stacks flawlessly)
# ==============================================================================
logo_img = Image.open("assets/logo.png")
logo_img = logo_img.resize((300, 299))
logo_photo = ImageTk.PhotoImage(logo_img)

logo_label = tk.Label(bg_label, image=logo_photo, bg='#B05938')
logo_label.place(relx=0, x=100, rely=0, y=40)
logo_label.image = logo_photo
logo_label.lift()  # Forces rendering to the front on macOS

# ==============================================================================
# 3. INTERFACE TEXT & WIDGETS
# ==============================================================================
# Centered Header Text
hello_label = tk.Label(root, text="What's your goal?", font=('Arial', 35, 'bold'), fg='#333333', bg='#B05938')
hello_label.place(relx=0.5, rely=0.3, anchor='center')
hello_label.lift()

welcome_label = tk.Label(root, text="Choose your role", font=('Arial', 26, 'bold'), fg='#666666', bg='#DCC8AD')
welcome_label.place(relx=0.5, rely=0.7, anchor='center')
welcome_label.lift()

# Button Frame Container
button_frame = tk.Frame(root, bg='#DCC8AD')
button_frame.place(relx=0.5, rely=0.9, y=-20, anchor='s')

# Inner Columns (Left & Right)
left_frame = tk.Frame(button_frame, bg='#DCC8AD')
left_frame.pack(side='left', padx=20)

right_frame = tk.Frame(button_frame, bg='#DCC8AD')
right_frame.pack(side='left', padx=20)

# --- EMPLOYER SECTION ---
employer_btn = tk.Button(left_frame, text="I AM AN EMPLOYER", font=('Arial', 16, 'bold'),
                         bg='#B85C2E', fg='white', padx=25, pady=15, cursor='hand2', bd=0,
                         command=on_employer_click)
employer_btn.pack(pady=(0, 5))

employer_desc = tk.Label(left_frame, text="I want to hire workers for my company.",
                         font=('Arial', 12), fg='#555555', bg='#DCC8AD')
employer_desc.pack()

# --- EMPLOYEE SECTION ---
employee_btn = tk.Button(right_frame, text="I AM AN EMPLOYEE", font=('Arial', 16, 'bold'),
                         bg='#B85C2E', fg='white', padx=25, pady=15, cursor='hand2', bd=0,
                         command=on_employee_click)
employee_btn.pack(pady=(0, 5))

employee_desc = tk.Label(right_frame, text="I am looking for jobs and opportunities.",
                         font=('Arial', 12), fg='#555555', bg='#DCC8AD')
employee_desc.pack()

# --- SIGN IN LINK ---
signin_link = tk.Button(root, text="Already have an account? Sign In", font=('Arial', 11),
                        fg='#B85C2E', bg='#DCC8AD', activebackground='white',
                        activeforeground='#8a451f', bd=0, cursor='hand2', command=go_to_signin)
signin_link.place(relx=1.0, rely=1.0, x=-20, y=-20, anchor='se')

# ==============================================================================
# EXECUTION
# ==============================================================================
root.mainloop()
```
