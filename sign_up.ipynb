```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import csv
import os

class User:
    """User class with encapsulated password"""
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password = password   
    
    def check_password(self, pwd):
        """Getter method to verify password"""
        return self.__password == pwd


def go_to_login():
    messagebox.showinfo("Info", "Login page will open soon.")
    
def sign_up():
    username = entry_username.get().strip()
    email = entry_email.get().strip()
    password = entry_password.get().strip()

    if not username or not email or not password:
        messagebox.showerror("Error", "All fields are required.")
        return

    file_path = "userinfo.csv"
    file_exists = os.path.isfile(file_path)

    if file_exists:
        with open(file_path, "r") as file:
            reader = csv.reader(file)
            for row in reader:
                if len(row) >= 2:
                    if row[0] == username:
                        messagebox.showerror("Error", "Username already taken.")
                        return
                    if row[1] == email:
                        messagebox.showerror("Error", "Email already registered.")
                        return

    # write new user
    with open(file_path, "a", newline="") as file:
        writer = csv.writer(file)

        if not file_exists:
            writer.writerow(["username", "email", "password"])

        writer.writerow([username, email, password])

    messagebox.showinfo("Success", f"Account created for {username}")

    entry_username.delete(0, tk.END)
    entry_email.delete(0, tk.END)
    entry_password.delete(0, tk.END)

    # Create new user and add to list
    new_user = User(username, email, password)
    messagebox.showinfo("Success", f"Account created for {username}! \n\nYou can now log in.")

def exit_app(event=None):
    """Exit the application (bound to Escape key)"""
    root.destroy()

# Create the main window
root = tk.Tk()
root.title("Hustle Hub")
root.attributes('-fullscreen', True)

# Bind Escape key to close the window
root.bind('<Escape>', exit_app)

# Load and resize image
bg_image = Image.open("assets/bg_image.png")
bg_image = bg_image.resize((root.winfo_screenwidth(), root.winfo_screenheight()))
bg_photo = ImageTk.PhotoImage(bg_image)

# Store reference on root to prevent garbage collection
root.bg_photo = bg_photo

# Create label using the stored reference
bg_label = tk.Label(root, image=root.bg_photo)
bg_label.place(x=0, y=0, relwidth=1, relheight=1)

logo_img = Image.open("assets/logo.png")
logo_img = logo_img.resize((300, 299))
logo_photo = ImageTk.PhotoImage(logo_img)
logo_label = tk.Label(root, image=logo_photo, bg='#B05938')
logo_label.place(relx=0, x=100, rely=0, y=40)
logo_label.image = logo_photo

#Create frame
frame = tk.Frame(root, width=460, height=542, bg="white")   # explicit size
frame.place(relx=1.0, x=-50, rely=1.0, y=-50, anchor='se')           # position it
frame.pack_propagate(False)

# Load and set background image for the frame
frame_bg_img = Image.open("assets/frame_bg.png")
frame_bg_img = frame_bg_img.resize((460, 542))
frame_bg_photo = ImageTk.PhotoImage(frame_bg_img)

frame_bg_label = tk.Label(frame, image=frame_bg_photo)
frame_bg_label.place(x=0, y=0, relwidth=1, relheight=1)
frame_bg_label.image = frame_bg_photo   # keep reference

welcome_label = tk.Label(root, text="Hello", font=("Kanit", 24, "bold"), fg="black", bg="#B05938")
welcome_label.place(in_=frame, relx=0.5, y=-70, anchor='s')

welcome_label = tk.Label(root, text="Welcome to Hustle Hub", font=("Kanit", 24), fg="black", bg="#B05938")

welcome_label.place(in_=frame, relx=0.5, y=-20, anchor='s')

# Heading
heading_label = tk.Label(frame, text="Create Account", font=("Arial", 22, "bold"), bg="#C0765B")
heading_label.pack(pady=(40, 20))

# Name label and entry 
label_username = tk.Label(frame, text="Username", font=("Arial", 15), bg="#C0765B")
label_username.pack(anchor='w', padx=20, pady=(0, 5))
entry_username = tk.Entry(frame, width=90)
entry_username.pack(anchor='w', padx=20, pady=(0, 15))

# Email label and entry
label_email = tk.Label(frame, text="Email Address", font=("Arial", 15), bg="#C0765B")
label_email.pack(anchor='w', padx=20, pady=(0, 5))
entry_email = tk.Entry(frame, width=90)
entry_email.pack(anchor='w', padx=20, pady=(0, 15))

# Password label and entry
label_password = tk.Label(frame, text="Password",font=("Arial", 15), bg="#C0765B")
label_password.pack(anchor='w', padx=20, pady=(0, 5))
entry_password = tk.Entry(frame, show="*", width=90)
entry_password.pack(anchor='w', padx=20, pady=(0, 30))

tagline = tk.Label(root, text="Hustle Hub, where every skill finds its worth.", font=("Segoe Script", 18), fg="black", bg="#DCC8AD")
tagline.place(x=30, rely=0.5, y=120)   # adjust y to sit below the logo

# Sign Up button 
btn_signup = tk.Button(frame, text="Sign Up", font=("Arial", 25, "bold"), command=sign_up, bg="#B05A39", cursor="hand2", fg="black")
btn_signup.pack(pady=(0, 0))
btn_signup.place(relx=0.5, rely=0.5, y=80, anchor='center', width=200, height=50)



# Login to Account button    
btn_login = tk.Button(frame, text="Go to login", font=("Arial", 12, "bold"), command=go_to_login, bg="#E4D1B9", cursor="hand2", fg="black")
btn_login.pack(pady=(0, 0))
btn_login.place(relx=0.5, rely=0.5, y=140, anchor='center')

# Run the application
root.mainloop()
```
