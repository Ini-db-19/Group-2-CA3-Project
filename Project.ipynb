```python
import tkinter as tk
from tkinter import messagebox

def exit_app(event=None):
    root.destroy()

class HustleHubApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Hustle Hub")
        self.root.attributes("-fullscreen", True)
        self.root.configure(bg='#f7f9fc')
        
        self.saved_jobs = []
        self.applied_jobs = []
        self.archived_jobs = []
        self.current_tab = "saved"
        
        self.create_sidebar()
        self.create_main_area()
    
    # ========== SIDEBAR ==========
    def create_sidebar(self):
        self.sidebar = tk.Frame(self.root, bg='#1E1E2E', width=250, relief='flat')
        self.sidebar.place(x=0, y=0, width=250, height=self.root.winfo_screenheight())
        self.sidebar.pack_propagate(False)
        
        header_frame = tk.Frame(self.sidebar, bg='#1E1E2E', height=100)
        header_frame.pack(fill='x', pady=(20, 10))
        logo_text = tk.Label(header_frame, text="Hustle Hub", 
                            font=('Segoe UI', 18, 'bold'),
                            fg='#FFB347', bg='#1E1E2E')
        logo_text.pack(pady=10)
        
        separator = tk.Frame(self.sidebar, bg='#3A3A4A', height=2)
        separator.pack(fill='x', padx=20, pady=5)
        
        self.make_sidebar_button("Home", "🏠", "Home", "Welcome to your home feed!")
        self.make_sidebar_button("My Jobs", "💼", "My Jobs", "View and manage your job applications.")
        self.make_sidebar_button("Messages", "💬", "Messages", "Check your conversations.")
        self.make_sidebar_button("Profile", "👤", "Profile", "Your personal information and settings.")
        
        bottom_frame = tk.Frame(self.sidebar, bg='#1E1E2E')
        bottom_frame.pack(side='bottom', fill='x', pady=20)
        logout_btn = tk.Button(bottom_frame, text="Logout", 
                              font=('Segoe UI', 10),
                              bg='#FF6B6B', fg='white', bd=0, 
                              padx=10, pady=5, cursor="hand2",
                              command=lambda: messagebox.showinfo("Logout", "You have been logged out."))
        logout_btn.pack(pady=5)
    
    def make_sidebar_button(self, text, icon, msg_title, msg_text):
        frame = tk.Frame(self.sidebar, bg='#1E1E2E')
        frame.pack(fill='x', padx=15, pady=5)
        label = tk.Label(frame, text=f"{icon}  {text}", 
                        font=('Segoe UI', 12),
                        fg='white', bg='#1E1E2E', 
                        anchor='w', cursor="hand2")
        label.pack(fill='x', padx=10, pady=8)
        label.bind('<Button-1>', lambda e: messagebox.showinfo(msg_title, msg_text))
    
    # ========== MY JOBS PAGE - SHIFTED GENTLY ==========
    def create_main_area(self):
        self.main_area = tk.Frame(self.root, bg='#f7f9fc')
        self.main_area.place(x=250, y=0, width=800, height=self.root.winfo_screenheight())
        self.main_area.pack_propagate(False)
        
        # Header and buttons: moved a bit to the LEFT (was 0.7, now 0.6)
        header_container = tk.Frame(self.main_area, bg='#f7f9fc')
        header_container.place(relx=0.65, rely=0.1, anchor='center')
        
        title = tk.Label(header_container, text="My Jobs", 
                        font=("Segoe UI", 28, "bold"),
                        bg='#f7f9fc', fg='#2c3e50')
        title.pack(pady=(0, 25))
        
        tab_frame = tk.Frame(header_container, bg='#f7f9fc')
        tab_frame.pack()
        
        self.saved_btn = tk.Button(tab_frame, text="Saved", 
                                   command=lambda: self.switch_tab("saved"),
                                   font=("Segoe UI", 11, "bold"), 
                                   width=12, padx=10, pady=5)
        self.saved_btn.pack(side='left', padx=8)
        
        self.applied_btn = tk.Button(tab_frame, text="Applied",
                                     command=lambda: self.switch_tab("applied"),
                                     font=("Segoe UI", 11, "bold"), 
                                     width=12, padx=10, pady=5)
        self.applied_btn.pack(side='left', padx=8)
        
        self.archived_btn = tk.Button(tab_frame, text="Archived",
                                      command=lambda: self.switch_tab("archived"),
                                      font=("Segoe UI", 11, "bold"), 
                                      width=12, padx=10, pady=5)
        self.archived_btn.pack(side='left', padx=8)
        
        # Content ("No jobs yet"): stays at relx=0.65 (unchanged)
        self.content_frame = tk.Frame(self.main_area, bg='#f7f9fc')
        self.content_frame.place(relx=0.65, rely=0.5, anchor='center')
        
        self.switch_tab("saved")
    
    def switch_tab(self, tab_name):
        self.current_tab = tab_name
        
        self.saved_btn.config(bg='#3498db' if tab_name == "saved" else '#ecf0f1',
                              fg='white' if tab_name == "saved" else '#34495e')
        self.applied_btn.config(bg='#3498db' if tab_name == "applied" else '#ecf0f1',
                                fg='white' if tab_name == "applied" else '#34495e')
        self.archived_btn.config(bg='#3498db' if tab_name == "archived" else '#ecf0f1',
                                 fg='white' if tab_name == "archived" else '#34495e')
        
        for widget in self.content_frame.winfo_children():
            widget.destroy()
        
        empty_frame = tk.Frame(self.content_frame, bg='#f7f9fc')
        empty_frame.pack()
        
        icon_label = tk.Label(empty_frame, text="📭", 
                              font=("Segoe UI", 48),
                              bg='#f7f9fc', fg='#bdc3c7')
        icon_label.pack(pady=(0, 15))
        
        empty_label = tk.Label(empty_frame, 
                               text="No jobs yet", 
                               font=("Segoe UI", 18),
                               bg='#f7f9fc', fg='#95a5a6')
        empty_label.pack()
        
        helper_text = tk.Label(empty_frame,
                               text="Start adding jobs to track your applications ✨",
                               font=("Segoe UI", 10),
                               bg='#f7f9fc', fg='#bdc3c7')
        helper_text.pack(pady=(10, 0))


if __name__ == "__main__":
    root = tk.Tk()
    root.bind('<Escape>', exit_app)
    app = HustleHubApp(root)
    root.mainloop()
```
