```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

def exit_app(event=None):
    root.destroy()
   
root = tk.Tk()
root.attributes('-fullscreen', True)
root.bind('<Escape>', exit_app)

# Get screen dimensions
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()

# Load and set background image
try:
    image = Image.open("assets/bg_image.png")
    image = image.resize((screen_width, screen_height))
    bg = ImageTk.PhotoImage(image)
    label = tk.Label(root, image=bg)
    label.place(x=0, y=0, relwidth=1, relheight=1)
except Exception as e:
    print(f"Error loading background image: {e}")
    # Fallback: set a background color
    root.configure(bg="#EDE3D6")

# Sidebar
sidebar = tk.Frame(root, bg='#1E1E2E', width=250, relief='flat')
sidebar.place(x=0, y=0, width=200, height=root.winfo_screenheight())
sidebar.pack_propagate(False)

header_frame = tk.Frame(sidebar, bg='#1E1E2E', height=100)
header_frame.pack(fill='x', pady=(20, 10))
logo_text = tk.Label(header_frame, text="Hustle Hub", font=('Segoe UI', 18, 'bold'),
                     fg='#FFB347', bg='#1E1E2E')
logo_text.pack(pady=10)
separator = tk.Frame(sidebar, bg='#3A3A4A', height=2)
separator.pack(fill='x', padx=20, pady=5)

def make_sidebar_button(text, icon, msg_title, msg_text):
    frame = tk.Frame(sidebar, bg='#1E1E2E')
    frame.pack(fill='x', padx=15, pady=5)
    label = tk.Label(frame, text=f"{icon}  {text}", font=('Segoe UI', 12),
                     fg='white', bg='#1E1E2E', anchor='w', cursor="hand2")
    label.pack(fill='x', padx=10, pady=8)
    label.bind('<Button-1>', lambda e: messagebox.showinfo(msg_title, msg_text))

make_sidebar_button("Home", "🏠", "Home", "Welcome to your home feed!")
make_sidebar_button("My Jobs", "💼", "My Jobs", "View and manage your job applications.")
make_sidebar_button("Messages", "💬", "Messages", "Check your conversations.")
make_sidebar_button("Profile", "👤", "Profile", "Your personal information and settings.")

bottom_frame = tk.Frame(sidebar, bg='#1E1E2E')
bottom_frame.pack(side='bottom', fill='x', pady=20)
logout_btn = tk.Button(bottom_frame, text="Logout", font=('Segoe UI', 10),
                       bg='#FF6B6B', fg='white', bd=0, padx=10, pady=5,
                       cursor="hand2", command=lambda: messagebox.showinfo("Logout", "You have been logged out."))
logout_btn.pack(pady=5)

#Search Bar Frame
searchbar = tk.Frame(root, bg="white")
searchbar.place(x=200, y=0, height=170, relwidth=1)

# Messages label
messages_label = tk.Label(root, text="Messages", bg="white", fg="#1F4E79", 
                          font=("Arial", 35, "bold"))
messages_label.place(relx=0.1, x=190, y=35, anchor="center")

# Search Icon
search_icon = Image.open("assets/search.png")
search_icon = search_icon.resize((25, 25))
search_photo = ImageTk.PhotoImage(search_icon)
search_label = tk.Label(searchbar, image=search_photo)
search_label.image = search_photo
search_label.place(relx=0.001, y=75, anchor="nw")

# Search bar
search_var = tk.StringVar()
search_bar = tk.Entry(searchbar, textvariable=search_var, font=("Arial", 12), width=60, bg="#E6E6E6", fg="black", 
                      insertbackground="black")
search_bar.place(relx=0.001, x=29, y=75, anchor="nw", width=1000, height=28)

def search_messages(event=None):
    """Search for messages"""
    query = search_var.get()
    if query:
        messagebox.showinfo("Search", f"Searching for: {query}")
    else:
        messagebox.showwarning("Empty Search", "Please enter a search term")

# Bind Enter key to search
search_bar.bind("<Return>", search_messages)

# Search button
search_button = tk.Button(searchbar, text="Search", command=search_messages,
                         bg="#1F4E79", fg="white", font=("Arial", 10))
search_button.place(x=1029, y=75, width=50, height=27.5)

def make_filter_button(parent, text, msg_title, msg_text):
    """Create filter buttons in the searchbar area"""
    btn = tk.Button(parent, text=text, font=('Segoe UI', 10),
                    bg='#F0F0F0', fg='#1F4E79', cursor="hand2",
                    bd=1, relief="solid", padx=15,
                    command=lambda: messagebox.showinfo(msg_title, msg_text))
    btn.pack(side="left", padx=5, pady=5)
    return btn

# Create filter buttons frame - MOVED FURTHER LEFT (from 200 to 50)
filter_button_frame = tk.Frame(searchbar, bg="white")
filter_button_frame.place(x=30, y=115)  # Changed from x=200 to x=50

make_filter_button(filter_button_frame, "All", "All Messages", "Showing all your messages")
make_filter_button(filter_button_frame, "Unread", "Unread Messages", "You have no unread messages")
make_filter_button(filter_button_frame, "Read", "Read Messages", "Showing messages you've read")
make_filter_button(filter_button_frame, "Starred", "Starred Messages", "Your starred messages will appear here")
make_filter_button(filter_button_frame, "Sent", "Sent Messages", "Showing messages you've sent")

root.mainloop()
```
